## **Preprocessing Data**

### hapus kolom ID jika ada.

Hapus Kolom 'ID': Menghapus kolom ID jika ada, karena tidak berguna untuk model.

```python
# Salin dataset agar tidak mengubah df asli
data = df.copy()

# Hapus kolom 'ID' jika ada
if 'ID' in data.columns:
    data.drop('ID', axis=1, inplace=True)
    print(" Kolom 'ID' telah dihapus.")
else:
    print(" Kolom 'ID' tidak ditemukan.")
```

### Isi missing values dengan modus

Isi Missing Values: Mengisi nilai yang hilang dengan modus (nilai yang paling sering muncul).

```python
# Mengisi missing value dengan modus (nilai yang paling sering muncul)
data.fillna(data.mode().iloc[0], inplace=True)

# Cek apakah masih ada missing value
print(" Missing value telah diisi.\n")
print("Jumlah nilai kosong setelah pengisian:")
print(data.isnull().sum())
```

### mengubah kategori ke angka

Encode Kolom Kategorikal: Mengubah data kategorikal menjadi angka dengan LabelEncoder.

```python
from sklearn.preprocessing import LabelEncoder

# Inisialisasi label encoder
label_encoder = LabelEncoder()

# Encode semua kolom bertipe object
for col in data.select_dtypes(include='object').columns:
    data[col] = label_encoder.fit_transform(data[col])
    print(f" Kolom '{col}' telah diencoding.")
```

### menormalkan fitur numerik (fitur target tidak ikut).

Standarisasi Fitur Numerik: Menormalkan fitur numerik dengan StandardScaler agar memiliki rata-rata 0 dan deviasi standar 1.

```python
from sklearn.preprocessing import StandardScaler

# Tentukan kolom target
target_column = 'Status'  # Ubah jika target kamu berbeda

# Pisahkan fitur dan target
X = data.drop(target_column, axis=1)
y = data[target_column]

# Standarisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Cek hasil
print(" Fitur telah distandarisasi.")
import numpy as np
print("Contoh data setelah scaling:")
print(np.round(X_scaled[:5], 2))
```

## **Pemodelan (Model Klasifikasi)**

1. Bagi data menjadi data latih dan data uji

2. Bangun model klasifikasi (di sini kita pakai Random Forest karena cocok untuk data tabular)

3. Latih model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# 1. Split data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 2. Inisialisasi model Random Forest
model = RandomForestClassifier(random_state=42)

# 3. Latih model dengan data latih
model.fit(X_train, y_train)

# 4. Cek akurasi pada data latih
train_score = model.score(X_train, y_train)
print(f" Model berhasil dilatih. Akurasi di data latih: {train_score:.2f}")
```

## **Evaluasi Model**

1. Prediksi data uji

2. Hitung metrik evaluasi: akurasi, precision, recall, f1-score

3. Tampilkan confusion matrix dan classification report

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Prediksi data uji
y_pred = model.predict(X_test)

# 2. Evaluasi metrik
print(" Evaluasi Model:")
print(f" Akurasi       : {accuracy_score(y_test, y_pred):.2f}")
print(f" Precision     : {precision_score(y_test, y_pred, average='weighted'):.2f}")
print(f" Recall        : {recall_score(y_test, y_pred, average='weighted'):.2f}")
print(f" F1 Score      : {f1_score(y_test, y_pred, average='weighted'):.2f}")

# 3. Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# 4. Classification report
print("\n Classification Report:")
print(classification_report(y_test, y_pred))
```
